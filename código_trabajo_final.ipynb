```python
import csv

# Datos base
usuarios = {}
reservas = {}
asientos = [['O' for _ in range(11)] for _ in range(11)]  # 121 asientos
peliculas = [
    {"id": 1, "nombre": "Interstellar", "dia": "Viernes", "horas": ["2pm", "4pm", "6pm"], "disponibles": [100, 5, 50]},
    {"id": 2, "nombre": "Oppenheimer", "dia": "Sabado", "horas": ["2pm", "4pm", "6pm"], "disponibles": [23, 44, 66]},
    {"id": 3, "nombre": "The Imitation Game", "dia": "Domingo", "horas": ["2pm", "4pm", "6pm"], "disponibles": [11, 0, 7]}
]
precios = {
    "Estudiantes": 7500,
    "Docentes": 10000,
    "Administrativos": 8500,
    "Oficiales internos": 7000,
    "Publico externo": 15000
}
admin_credenciales = {"admin": "1234"}

# Validaciones
def validar_nombre(nombre):
    return nombre.isalpha() and len(nombre) >= 3

def validar_documento(doc):
    return doc.isdigit() and 3 <= len(doc) <= 15

def registrar_usuario():
    nombre = input("Nombre: ")
    apellido = input("Apellido: ")
    documento = input("Documento: ")
    tipo = input("Tipo de vínculo (Estudiantes, Docentes, Administrativos, Oficiales internos, Publico externo): ")

    errores = []
    if not validar_nombre(nombre): errores.append("Nombre inválido.")
    if not validar_nombre(apellido): errores.append("Apellido inválido.")
    if not validar_documento(documento): errores.append("Documento inválido.")
    if tipo not in precios: errores.append("Tipo de vínculo no válido.")

    if errores:
        print("Errores encontrados:")
        for e in errores:
            print("-", e)
        return

    usuarios[documento] = {"nombre": nombre, "apellido": apellido, "tipo": tipo}
    print("Usuario registrado exitosamente.")

def mostrar_cartelera():
    print("\nCARTELERA DE CINE")
    for peli in peliculas:
        for i, hora in enumerate(peli["horas"]):
            print(f"{peli['id']} - {peli['dia']} {hora} - {peli['nombre']} - Disponibles: {peli['disponibles'][i]}")

def reservar_asiento():
    doc = input("Documento del usuario: ")
    if doc not in usuarios:
        print("Usuario no registrado.")
        return

    mostrar_cartelera()
    id_peli = int(input("ID de película: "))
    hora = input("Hora (ej. 2pm): ")
    fila = int(input("Fila (0-10): "))
    col = int(input("Columna (0-10): "))

    if asientos[fila][col] == 'X':
        print("Asiento ocupado.")
        return

    asientos[fila][col] = 'X'
    reservas[doc] = {"pelicula": id_peli, "hora": hora, "asiento": (fila, col)}
    precio = precios[usuarios[doc]["tipo"]]
    print(f"Reserva confirmada. Precio: ${precio}")
    print("Factura generada.")

def cancelar_reserva():
    doc = input("Documento del usuario: ")
    if doc not in reservas:
        print("No hay reservas activas.")
        return

    fila, col = reservas[doc]["asiento"]
    asientos[fila][col] = 'O'
    del reservas[doc]
    print("Reserva cancelada.")

def administrador():
    user = input("Usuario admin: ")
    pwd = input("Contraseña: ")
    if admin_credenciales.get(user) != pwd:
        print("Credenciales incorrectas.")
        return

    print("\n--- Reportes Administrativos ---")
    print("Total reservas:", len(reservas))
    print("Total usuarios:", len(usuarios))
    total_pago = sum(precios[usuarios[doc]["tipo"]] for doc in reservas)
    print("Total pago realizado:", total_pago)

def menu():
    while True:
        print("\n--- Luces, camara, UdeA ---")
        print("1. Registrar Usuario")
        print("2. Registrar Reserva")
        print("3. Cancelar Reserva")
        print("4. Consultar Funciones Fin de Semana")
        print("5. Administrador")
        print("6. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            registrar_usuario()
        elif opcion == "2":
            reservar_asiento()
        elif opcion == "3":
            cancelar_reserva()
        elif opcion == "4":
            mostrar_cartelera()
        elif opcion == "5":
            administrador()
        elif opcion == "6":
            break
        else:
            print("Opción inválida.")
```

```python
menu()
```
